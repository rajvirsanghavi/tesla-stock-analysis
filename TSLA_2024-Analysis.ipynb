```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download Tesla data - last 1 year
tesla = yf.download("TSLA", start="2024-01-01", end="2025-01-01")

# Preview the data
print(tesla.head())
print(f"\nTotal trading days: {len(tesla)}")
```

```python
# Plot Tesla's closing price
plt.figure(figsize=(12, 6))
plt.plot(tesla['Close'], color='red', linewidth=1.5)
plt.title("Tesla (TSLA) Stock Price - 2024", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Calculate 50-day and 200-day moving averages
tesla['MA50'] = tesla['Close'].rolling(window=50).mean()
tesla['MA200'] = tesla['Close'].rolling(window=200).mean()

# Plot everything together
plt.figure(figsize=(12, 6))
plt.plot(tesla['Close'], color='red', linewidth=1.5, label='TSLA Price', alpha=0.7)
plt.plot(tesla['MA50'], color='blue', linewidth=2, label='50-Day MA')
plt.plot(tesla['MA200'], color='orange', linewidth=2, label='200-Day MA')

plt.title("Tesla (TSLA) - Price & Moving Averages 2024", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Re-download data
tesla = yf.download("TSLA", start="2024-01-01", end="2025-01-01", auto_adjust=True)

# Calculate everything
tesla['Daily_Return'] = tesla['Close'].pct_change()

# Plot
plt.figure(figsize=(12, 4))
plt.plot(tesla['Daily_Return'], color='purple', linewidth=0.8, alpha=0.7)
plt.axhline(y=0, color='black', linewidth=1)
plt.title("Tesla (TSLA) - Daily Returns 2024", fontsize=16)
plt.xlabel("Date")
plt.ylabel("Daily Return")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Print stats
avg_return = float(tesla['Daily_Return'].mean()) * 100
volatility = float(tesla['Daily_Return'].std()) * np.sqrt(252) * 100
total_return = float((tesla['Close'].iloc[-1] / tesla['Close'].iloc[0]) - 1) * 100

print(f"Average Daily Return: {avg_return:.3f}%")
print(f"Annual Volatility:    {volatility:.1f}%")
print(f"Total 2024 Return:    {total_return:.1f}%")
```

```python
# Save all charts as images for your portfolio
fig1, ax1 = plt.subplots(figsize=(12, 6))
ax1.plot(tesla['Close'], color='red', linewidth=1.5)
ax1.set_title("Tesla (TSLA) Stock Price - 2024", fontsize=16)
ax1.set_xlabel("Date")
ax1.set_ylabel("Price (USD)")
ax1.grid(True, alpha=0.3)
fig1.savefig("tesla_price.png", dpi=150, bbox_inches='tight')

fig2, ax2 = plt.subplots(figsize=(12, 4))
ax2.plot(tesla['Daily_Return'], color='purple', linewidth=0.8, alpha=0.7)
ax2.axhline(y=0, color='black', linewidth=1)
ax2.set_title("Tesla (TSLA) - Daily Returns 2024", fontsize=16)
ax2.grid(True, alpha=0.3)
fig2.savefig("tesla_returns.png", dpi=150, bbox_inches='tight')

print("Charts saved!")
```

```python
plt.figure(figsize=(10, 6))
plt.hist(tesla['Daily_Return'].dropna(), bins=50, color='steelblue',
         edgecolor='white', linewidth=0.5)

# Add vertical lines for key stats
plt.axvline(float(tesla['Daily_Return'].mean()), color='green',
            linewidth=2, linestyle='--', label=f'Mean: {avg_return:.3f}%')
plt.axvline(float(tesla['Daily_Return'].mean()) + float(tesla['Daily_Return'].std()),
            color='orange', linewidth=2, linestyle='--', label='+1 Std Dev')
plt.axvline(float(tesla['Daily_Return'].mean()) - float(tesla['Daily_Return'].std()),
            color='red', linewidth=2, linestyle='--', label='-1 Std Dev')

plt.title("Tesla (TSLA) - Distribution of Daily Returns 2024", fontsize=16)
plt.xlabel("Daily Return")
plt.ylabel("Frequency (Number of Days)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("tesla_histogram.png", dpi=150, bbox_inches='tight')
plt.show()

print("Histogram saved!")
```
